```python
def Print_info(name: str, iub_id: int, cgpa: float,age:int,d:str):
        data = {
        "name": name,
        "id": iub_id,
        "cgpa": cgpa,
        'age':age,
        'department': d
        }
        for key,value in data.items():
            print(f"{key}: {value}")
        print('----------------------')
```

```python
class Student:
    def __init__(self,name:str,age:int):
        self.name = name
        self.age = age

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}", end=' ')
    
class CLUB(Student):
    def __init__(self,name:str,age:int,member:str):
        super().__init__(name,age)
        self.section = member

    def display(self):
        super().display()
        print(f"Member Of: {self.section}")
fahim = Student('Fahim', 25)
sabbir = CLUB('Sabbir', 22, 'JUKTI')
fahim.display()
sabbir.display()
```
